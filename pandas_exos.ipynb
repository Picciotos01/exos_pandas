```python
pourcentage_survivants_par_sexe = df.groupby('sex')['survived'].mean() * 100
pourcentage_survivants_par_sexe
```

```python
classe_max_survivants = df.groupby('class')['survived'].sum().idxmax()
classe_max_survivants
```

```python
moyenne_tarifs_survivants = df[df['survived'] == 1]['fare'].mean()
moyenne_tarifs_non_survivants = df[df['survived'] == 0]['fare'].mean()
print("Moyenne des tarifs des survivants :", moyenne_tarifs_survivants)
print("Moyenne des tarifs des non-survivants :", moyenne_tarifs_non_survivants)
```

```python
tarif_median = df['fare'].median()
passagers_tarif_superieur_median = df[df['fare'] > tarif_median]
```

```python
df['title'] = df['name'].str.extract('([A-Za-z]+)\.')
print(df['title'].value_counts())
```

```python
categories = ['baby', 'kid', 'adolescent', 'adult']
df['group_age'] = pd.cut(df['age'], bins=[0, 2, 11, 17, float('inf')], labels=categories)
df
```

```python
passagers_mineurs = df[(df['age'] < 18) & ((df['sibsp'] != 0) & (df['parch'] != 0))]
print("Passagers mineurs avec SibSp ou Parch diffÃ©rent de zÃ©ro :")
print(passagers_mineurs.head())
```

```python
correlation_age_tarif = df['age'].corr(df['fare'])
print("CorrÃ©lation entre l'Ã¢ge et le tarif payÃ© par les passagers :", correlation_age_tarif)
```
